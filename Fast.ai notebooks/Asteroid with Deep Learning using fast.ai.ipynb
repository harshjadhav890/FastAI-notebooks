```python
import fastai
from fastai.tabular.all import *

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Asteroid_Updated.csv (1).zip (Unzipped Files)/Asteroid_Updated.csv', low_memory=False)
```

```python
df = df.dropna(subset=["diameter", "rot_per"])
df = df[pd.to_numeric(df['diameter'], errors='coerce').notnull()]
df = df.interpolate()
df = df.round(decimals=5)
```

```python
cont_names = ['a', 'e', 'i', 'om', 'w', 'q', 'ad', 'per_y', 'data_arc', 'n_obs_used',
   'H', 'diameter', 'albedo', 'rot_per', 'moid','n', 'per', 'ma', 'GM', 'IR', 'BV', 'UB', 'G']
cat_names=['spec_B', 'spec_T', 'class']
df[cont_names]= df[cont_names].astype(float)
df = df[(cont_names + cat_names)]
```

```python
df.head()
```

```python
dls = TabularDataLoaders.from_df(df, procs = [Categorify, FillMissing, Normalize], cat_names=['spec_B', 'spec_T', 'class'], 
                           cont_names=['a', 'e', 'i', 'om', 'w', 'q', 'ad', 'per_y', 'data_arc', 'n_obs_used', 'H', 'diameter', 
                                       'albedo', 'rot_per', 'moid','n', 'per', 'ma', 'GM', 'IR', 'BV', 'UB', 'G'],
                           y_names='diameter')
```

```python
learn = tabular_learner(dls, metrics=rmse)
learn.fit_one_cycle(3)
```
